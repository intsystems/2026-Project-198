Плагиат изображений является серьёзной проблемой в журналистике, академических публикациях, искусстве и дизайне. Злоумышленники маскируют заимствования, применяя к исходному изображению геометрические преобразования: повороты, отражения, сдвиги, масштабирование, а также добавляют шум, водяные знаки и цветовые искажения. Стандартные энкодеры на основе Vision Transformer (ViT) [@dosovitskiy2021image] не обладают инвариантностью относительно указанных преобразований по построению и вынуждены полагаться на аугментацию данных при обучении, что даёт лишь приближённую устойчивость и увеличивает вычислительные затраты. Недавние исследования показали, что современные модели не справляются с задачами пространственного и геометрического рассуждения [@chen2024spatialvlm], что свидетельствует о недостаточности масштабирования моделей для решения проблемы инвариантности. Альтернативный подход - встроить геометрическую симметрию непосредственно в архитектуру энкодера с помощью эквивариантных нейронных сетей [@cohen2016group; @bronstein2021geometric]. Функция $\varphi$ называется $G$-эквивариантной относительно группы преобразований $G$, если применение действия элемента $g \in G$ к входу приводит к фиксированному, заранее определённому действию на выход: $\varphi(\rho_X(g) \cdot x) = \rho_Y(g) \cdot \varphi(x)$, где $\rho_X$ и $\rho_Y$ - представления группы $G$ во входном и выходном пространствах соответственно. Инвариантность является частным случаем эквивариантности, при котором действие на выход тривиально и выход не изменяется: $\varphi(\rho_X(g) \cdot x) = \varphi(x)$. Для задачи детекции плагиата $G$-инвариантный энкодер гарантирует, что эмбеддинг изображения не изменится при любом преобразовании из группы $G$, устраняя необходимость в соответствующих аугментациях.

Цель исследования - оценить влияние геометрического индуктивного смещения в архитектуре энкодера на качество детекции плагиата изображений и ответить на вопрос: даёт ли встроенная инвариантность эмбеддингов относительно группы симметрий $G$ преимущество перед аугментацией данных при обучении нейронной сети. Рассматриваются три группы преобразований: группа диэдра $D_4$ порядка 8 (повороты на кратные $90°$ и отражения - типичные манипуляции при плагиате), группа целочисленных трансляций $\mathbb{Z}^2$ (циклические сдвиги изображения) и специальная евклидова группа $SE(2)$ (непрерывные повороты плоскости на произвольный угол и сдвиги). В качестве детектора плагиата используется сиамская сеть [@bromley1993signature] - архитектура с двумя ветвями и общими весами, отображающая пару изображений в оценку сходства. Сравниваются четыре варианта энкодера. Стандартный ViT-L/16 [@dosovitskiy2021image] - без каких-либо гарантий инвариантности, служит базовым методом. Octic ViT [@nordstrom2025octic] - энкодер, слои которого спроектированы так, чтобы быть эквивариантными относительно $D_4$; на последнем этапе применяется инвариантный пулинг, и выходной эмбеддинг не меняется при повороте или отражении входного изображения. Shift-Equivariant ViT [@rojasgomez2024shift] - энкодер, в котором модули токенизации, самовнимания и понижения размерности перепроектированы для обеспечения эквивариантности относительно циклических сдвигов. Harmformer [@karella2024harmformer] - энкодер, комбинирующий круговые гармонические фильтры с трансформерными блоками для достижения эквивариантности относительно непрерывных поворотов и сдвигов, в отличие от Octic ViT, который ограничен углами, кратными $90°$.
